# **Data Preprocessing**

```python
#including necessary packages
from itertools import islice
from google.colab import drive
import pandas as pd
import os

#reading the csv files from directory
drive.mount('/content/drive',force_remount=True)
path_of_the_directory= 'drive/My Drive/CallsSMS Data'
dataframes_list1 = []

for filename in os.listdir(path_of_the_directory):
    f = os.path.join(path_of_the_directory,filename)
    if os.path.isfile(f):
        dataset=pd.read_csv(f)

        #Date and Time values are converted datetime format.
        dataset['Date'] = dataset['Date'].astype('datetime64[ns]')
        dataset['Time'] = dataset['Time'].astype('string')
        dataset['Time'] = pd.to_datetime(dataset['Time'],format= '%H:%M:%S' )

        #Day is extracted from date and value of hour is extracted from time.
        dataset["Weekday"] = dataset["Date"].dt.dayofweek
        dataset["Hour"] = dataset["Time"].dt.hour

        #Date and Time columns are dropped as corresponding day and hour has been extracted
        columns = ["Date","Time"]
        data_set = dataset.drop(columns, axis=1)

        #Categorical values are converted to numerical values for Event and Type.
        dataset_final=pd.get_dummies(data_set, columns=["Event","Type"])

        #individual dataframes for all the files are added to a list for later use
        dataframes_list1.append(dataset_final)

#displaying dataframes for all csv files
for dataset in dataframes_list1:
  display(dataset)
```

**dataframes are checked for null values**

```python
length=len(dataframes_list1)

for i in range(0,length):
  data_set=dataframes_list1[i]
  print("Null values for ",i, " " , data_set.isna().sum())
```

**As it can be seen that only number column have null values. Those rows are dropped in the following code**

```python
length=len(dataframes_list1)
dataframes_list=[]
for i in range(0,length):
  data_set=dataframes_list1[i]
  data_set=data_set.dropna(subset = ['Number'])
  dataframes_list.append(data_set)
```

# **Model Implementation**

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN
from sklearn.model_selection import train_test_split
```

### **Simple implementation based on frequency**
 In this approach

1. Data is split into train and test

2. Each record from test file is selected.

3. Weekday and hour are selected for each test row.

4. All examples from train dataset that have same hour and day value are selected.

5. Frequency is calculated for each user number i.e how many calls or sms were made to a particular user on that particular day and that particular time.

6. Top 8 user numbers and Event i.e call/sms are listed for each test row.

7. For accuracy, user number and event are selected for each test row and compared with the 8 predicted users. If atleast one of the predicted user and event is same as the original, the prediction is considered right. Else it is considered wrong.





```python
accuracies=[]

for i in range(0,length):
  dataset_final=dataframes_list[i]

  #splitting data into test and train
  X_train, X_test, = train_test_split(dataset_final, test_size = 0.05, random_state = 0)
  z=X_test.shape[0]
  right=0
  wrong=0

  for v in range(0,z):
    #selecting weekday and hour from each point
    wday=X_test.iloc[v]
    wday=wday["Weekday"]
    hr=X_test.iloc[v]
    hr=hr["Hour"]

    #getting points from train data that have day and hour similar to test point
    dataset_pred=X_train[X_train["Weekday"]==wday]
    dataset_pred=dataset_pred[dataset_pred["Hour"]==hr]
    df=dataset_pred.groupby(["Number", "Event_Call"]).size().reset_index(name="Frequency")
    df2=df.sort_values(['Frequency'], ascending=False)
    df2['Event_Call'] = df['Event_Call'].replace(0,'SMS')
    df2['Event_Call'] = df2['Event_Call'].replace(1,'Call')
    df2=df2.drop('Frequency', axis=1)

    #predicting top 8
    print("Top 8 Most likely contacts for day " + str(wday) + "and hour " + str(hr))
    print(df2.head(8))

    check=df2['Number'].to_numpy()

    #checking which predictions are correct and which are not
    if X_test.iloc[v].values[0:1] in check:
      right+=1
    else:
      wrong+=1


  print("Right predictions:" , right)
  print("wrong predictions:", wrong)
  accuracy= (right/(right+wrong))
  accuracies.append(accuracy)
  print("Accuracy for this implementation is" ,  accuracy*100, "%")
```

### **Accuracy**
Accuracy of all files is averaged for calculating overall accuracy.

```python
Overall_acc=sum(accuracies)/len(accuracies)
print("Average accuracy for this implementation is" , Overall_acc*100, "%")
```

### **Implementing DBSCAN**
1. For each file, data is splitted to test and split.

2. DBSCAN is implemented for each dataset, and clusters are formed.

3. For each test point in test dataset, it is compared to core points formed from clustering.

4. Point is assigned to the cluster where core point has smallest distance from the point.

5. In that particular cluster, distance is calculated to each point of that cluster.

6. Nearest 8 points are found, and user and number of that point is assigned to the test point.

9. For accuracy, user number and event are selected for each test row and compared with the 8 predicted users. If atleast one of the predicted user and event is same as the original, the prediction is considered right. Else it is considered wrong.

```python
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
from sklearn.cluster import DBSCAN

#creating lists to store datasets for each file
dbscan_list=[]
X_test_list=[]

for i in range(0,length):

  dataset_final=dataframes_list[i]


  #splitting data into test and train
  X_train, X_test, = train_test_split(dataset_final, test_size = 0.05, random_state = 0)

  X_test_list.append(X_test)
  right=0
  wrong=0

  #DBSCAN model implementation
  clustering = DBSCAN(eps=5, min_samples=5).fit(X_train)

  DBSCAN_dataset = X_train.copy()
  DBSCAN_dataset.loc[:,'Cluster'] = clustering.labels_
  DBSCAN_dataset.Cluster.value_counts().to_frame()


  outliers = DBSCAN_dataset[DBSCAN_dataset['Cluster']==-1]
  dbscan_list.append(DBSCAN_dataset)

  #plotting clusters
  plt.figure(figsize=(20,10))

  sns.scatterplot(DBSCAN_dataset["Hour"], DBSCAN_dataset["Number"],

                  data=DBSCAN_dataset[DBSCAN_dataset['Cluster']!=-1],

                  hue='Cluster', palette='Set1', legend='full', s=200)


  plt.show()
```

```python
print(len(dbscan_list))
print(len(X_test_list))
```

```python
from scipy import spatial

#initializing array to store accuracies
acc_dbscan=[]
le=len(dbscan_list)


for i in range(0,le):

  #getting test and train datasets
  test_data=X_test_list[i]
  DBSCAN_dataset=dbscan_list[i]


  #initializing variables for predicting correct and incorrect
  correct=0
  incorrect=0
  z=len(test_data)

  for v in range(0,z):

    z=DBSCAN_dataset.shape[0]
    arr = np.zeros((z,4)) #Number,Similarity,EventCall,Cluster

    #Comparing each test sample to all core points in order to assign it to a cluster
    for i in range(0,DBSCAN_dataset.shape[0]):
      sim=1-spatial.distance.cosine(DBSCAN_dataset.iloc[i].values[:-1], test_data.iloc[v].values)

      arr[i][0]=DBSCAN_dataset["Number"].iloc[i]
      arr[i][1]=sim
      arr[i][2]=DBSCAN_dataset["Event_Call"].iloc[i]
      arr[i][3]=DBSCAN_dataset["Cluster"].iloc[i]

    sort_arr=sorted(arr, key=lambda arr:arr[1])
    sort_arr.reverse()

    #picking up most similar points from that cluster to offer them as predictions
    data=DBSCAN_dataset[DBSCAN_dataset['Cluster']==sort_arr[1][3]]
    c=data.shape[0]
    arr = np.zeros((c,4)) #Number,Similarity,EventCall
    for i in range(0,data.shape[0]):
      sim=1-spatial.distance.cosine(data.iloc[i].values[:-1], test_data.iloc[v].values)
      arr[i][0]=data["Number"].iloc[i]
      arr[i][1]=sim
      arr[i][2]=data["Event_Call"].iloc[i]
      arr[i][3]=data["Cluster"].iloc[i]

    sort_arr=sorted(arr, key=lambda arr:arr[1])
    sort_arr.reverse()
    df=pd.DataFrame(sort_arr,columns = ["UserNumber", "Similarity", "Event","Cluster"])
    df = df.drop_duplicates('UserNumber', keep='first')
    df['Event'] = df['Event'].replace(0.0,'SMS')
    df['Event'] = df['Event'].replace(1.0,'Call')
    print("Predictions for day" + str(test_data.iloc[v].values[1:2]) + "and time" + str(test_data.iloc[v].values[2:3]))
    print(df.head(7))


    check=df['UserNumber'].to_numpy()
    if test_data.iloc[v].values[0:1] in check:
      correct+=1
    else:
      incorrect+=1

  print("Right predictions:" , correct)
  print("wrong predictions:", incorrect)
  acc_dbscan.append(correct/(correct+incorrect))

```

```python
Overall_acc_method2=sum(acc_dbscan)/len(acc_dbscan)
print("Average accuracy for this implementation is" , Overall_acc_method2*100, "%")
```

### **Random Forest**
Random forest was implemeneted as a supervised learning algorithm to check its accuracy on the dataset to predict one call / sms at a particular time.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report

acc_RF=[]

#implementing loop for each file
for i in range(0,len(dataframes_list)):

  dataset_final=dataframes_list[i]

  X= dataset_final.drop('Number', axis = 1)
  Y=dataset_final["Number"]

  #splitting into test and train.
  #"number" is considered as target variable
  X_train, X_test, y_train, y_test = train_test_split(
            X, Y, test_size = 0.05, random_state = 42)

  #model implementation
  clf =RandomForestClassifier()

  #training
  clf.fit(X_train, y_train)

  #predicting
  predictions = clf.predict(X_test)

  #accuracy for each file
  accuracy = accuracy_score(y_test, predictions)
  acc_RF.append(accuracy)
  X_test["Predicted"]=predictions

  #predictions are printed against original target values
  horizontal_concat = pd.concat([X_test, y_test], axis=1)

  print(horizontal_concat)
  print(classification_report(y_test, predictions))
```

```python
#overall accuracy for 1 predicted user using random forest

Overall_acc_RF=sum(acc_RF)/len(acc_RF)
print("Average accuracy for this implementation is" , Overall_acc_RF*100, "%")
```
